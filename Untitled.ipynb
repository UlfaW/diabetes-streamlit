Import Library

```python
import pandas as pd
import numpy as np
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

Dataset

```python
df = pd.read_csv('diabetes (3).csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

Transformasi Data

```python
kolom = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for i in kolom:
    df[i] = df[i].replace(0, df[i].mean())
```

Data Cleaning

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print("Jumlah Data Duplikat :", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

EDA (EXPLORATORY DATA ANALYSIS)

```python
print(df.describe())
```

```python
print(df['Outcome'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Outcome'].value_counts().plot(kind='bar')

plt.title('Jumlah Pasien Diabetes')
plt.xlabel('Outcome')
plt.ylabel('Jumlah')

plt.show()
```

Variabel X dan Y

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

Standar Scaler

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

Split Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

Training Model

```python
model = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=4,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)

model.fit(X_train, y_train)
```

Evaluari Akurasi

```python
y_pred = model.predict(X_test)

akurasi = accuracy_score(y_test, y_pred)

print("Akurasi Model :", akurasi * 100, "%")
```

```python
print(classification_report(y_test, y_pred))
```

```python
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print("Training Accuracy :", train_score * 100)
print("Testing Accuracy  :", test_score * 100)
```

Prediksi Manual

```python
data_baru = [[2,138,62,35,0,33.6,0.127,47]]

data_baru = scaler.transform(data_baru)

hasil = model.predict(data_baru)

if hasil[0] == 1:
    print("Pasien Terindikasi Diabetes")
else:
    print("Pasien Tidak Diabetes")
```

Simpan Model

```python
pickle.dump(model, open('model_diabetes.pkl', 'wb'))
pickle.dump(scaler, open('scaler.pkl', 'wb'))

print("Model berhasil disimpan")
```
